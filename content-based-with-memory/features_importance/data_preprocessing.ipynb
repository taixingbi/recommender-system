```python
import pandas as pd
import numpy as np
import random
"""
1. read all possible features
2. convert 't' and 'f' to 1 and 0
3. convert 'location' and 'text' to value (in future)
4. save as X.csv for training data
"""
class data_processing():
    def __init__(self, path_nsproperties_apt):
        self.path_nsproperties_apt= path_nsproperties_apt

    def apt_feature(self):
        def read_apt_features(path_nsproperties_apt):
            df= pd.read_csv(path_nsproperties_apt, index_col = False)
            features= ['is_rental', 'is_commercial', 'rent', 'price', 'num_bedrooms', 'area', 'num_rooms', 
                       'num_bathrooms', 'pets', 'is_new_development','common_charges', 'is_furnished']
            df= df[features]
            
            print(path_nsproperties_apt)
            print(features)
            return df
        
        def tf2binary(df):
            #convert "t" and "f" to number 1 and 0
            def f(row):
                features= ['is_rental', 'is_commercial','is_new_development', 'is_furnished']
                for col in features:
                    if row[col]=='t': row[col]= 1
                    if row[col]=='f': row[col]= 0
                return row
            df= df.apply(f, axis=1)
            return df
        
        def save_apt_features(df):
            path_save= 'data/X.csv'
            df.to_csv(path_save, index=False)
            print(path_save)
            
        df= read_apt_features(self.path_nsproperties_apt)
        df= tf2binary(df)
        save_apt_features(df)
        
        print(df.shape)
        return df
    
path_nsproperties_apt= "../pipenv/data/nsproperties_apt_exclusive_is_available_list_on_web_nyc.csv" 
df= data_processing(path_nsproperties_apt).apt_feature()
df.head()
```
